```python
from csv import reader
```

```python
dataset = list(reader(open("covid_19_data.csv")))
```

```python
header = dataset[0]
dataset = dataset[1:]
```

```python
header.index('Country/Region')
```

```python
country = [] # ["Mainland China"]

for i in dataset:
    if i[3] not in country:
        country.append(i[3])
```

```python
print(len(country))
```

```python
header
```

```python
country_china = {"Confirmed":0,"Deaths":0}
for i in dataset:
    if "Mainland China" == i[3]:
        country_china["Confirmed"] += float(i[-3])
        country_china["Deaths"] += float(i[-2])
```

```python
country_china
```

```python
analysis = [["country_name","total_confirmed","total_deaths","total_recovered"]]
analysis
```

```python
for i in country[200:]: # i : Hong Kong
    total_confirmed = 0
    total_deaths  = 0
    total_recovered = 0
    for j in dataset:
        if i == j[3]:
            total_confirmed += float(j[-3])
            total_deaths += float(j[-2])
            total_recovered += float(j[-1])
    analysis.append([i, total_confirmed, total_deaths, total_recovered])
#     print([i, total_confirmed, total_deaths, total_recovered])
#             print(i,float(j[-3]),float(j[-2]),float(j[-1]))
```

```python
dataset_2020 = []

for i in dataset:
    if i[1].year == 2020:
          dataset_2020.append(i)
```

```python
month_analysis = [["month_name","deaths"]]
```

```python

for i in range(1,13,1): # 2
    deaths = 0 # 0
    for j in dataset_2020: # [a,1990-3-2]
        if j[1].month == i: # 2 == 2
            date = j[1] # 1990-3-2
#             print(date)
            deaths += float(j[-2])
    month_analysis.append([date.strftime("%b"), deaths])
```
